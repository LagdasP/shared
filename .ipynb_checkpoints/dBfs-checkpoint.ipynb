```python
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import signal
from scipy.io import wavfile
```

```python
# Read the wave files
X = wavfile.read('YourPATH here/Test Tone 16int.wav')
Y = wavfile.read('YourPATH here/Test Tone 32int.wav')
Z = wavfile.read('YourPATH here/Test Tone 32float.wav')

#download files here: https://drive.google.com/drive/folders/1i6zjynHi-9ppfZ2k0-f151vLQ09XBZqV?usp=sharing

#Extract the audio infromation
int16_0dBfs = X[1]
int32_0dBfs = Y[1]
float32_0dBfs = Z[1]
```

```python
# This is a simple function that calculates dB
# Reference is measured as the peak to peak range
def dB(v, reference):
    return format(20*np.log10(v/(reference*.5)), ".3f")
```

```python
#Plot the first 100 samples from both audio files
#pay attention to the amplitude axis values

plt.figure(figsize = [18,6])
plt.subplot(1,3,1)
plt.plot(int16_0dBfs[0:100]);
plt.title("16int @0dBfs | Maximum level: " + str(max(int16_0dBfs)))

plt.subplot(1,3,2)
plt.plot(int32_0dBfs[0:100]);
plt.title("32int @0dBfs | Maximum level: " + str(max(int32_0dBfs)))

plt.subplot(1,3,3)
plt.plot(float32_0dBfs[0:100]);
plt.title("32float @0dBfs | Maximum level: " + str(max(float32_0dBfs)))

plt.show()
```

```python
print("Test Tone 16int.wav: ", dB(max(int16_0dBfs), (2**16)),"dB")
print("Test Tone 32int.wav: ", dB(max(int32_0dBfs), (2**32)),"dB")
print("Test Tone 32float.wav: ", dB(max(float32_0dBfs), (2)),"dB")
```
